```python
#import sqlalchemy to use sql db here
from sqlalchemy import create_engine 
#import pandas for data manupulation
import pandas as pd
#PostgreSQL database adapter for Python 
import psycopg2
#for config 
import json
# Matplotlib
import matplotlib.pyplot as plt
# NumPy
import numpy as np
```

```python
#open the json dict
with open('config.json', 'r') as c:
    cfg = json.load(c)
#debug
#print(cfg)
```

```python
#create a conn string
conn_str = "host={} dbname={} user={} password={}".format(
    cfg['host'], cfg['dbname'], cfg['user'], cfg['password'])                                                       
#debug 
#print(conn_str)
```

```python
#conn takes that same dict ( if it's correctly configured, as kwargs)
conn = psycopg2.connect(**cfg)
```

```python
#read the sql into a dataframe
salaries_df = pd.read_sql('select * from salaries', conn)
salaries_df.head()
```

```python
titles_df = pd.read_sql('select * from titles', conn)
titles_df.head()
```

```python
salary_bands = pd.merge(salaries_df, titles_df, on="emp_no", how="inner")
salary_bands.head()


```

```python
salaries_per_title = salary_bands.groupby('title').mean()
salaries_per_title
```

```python

df = salaries_per_title.drop(columns='emp_no')
#need to do this for the graph
df_bar = df.reset_index()
df_bar
```

```python


x_axis = df_bar['title']
ticks = np.arange(len(x_axis))
y_axis = df_bar['salary']
 
# Create Bar Chart Based on Above Data
plt.bar(x_axis, y_axis, align="center", alpha=0.8)

# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="50")

plt.ylabel("Salary in Dollars")
# plt.xlabel("Title")
plt.title("Salary/Title at Pewlett Hackard 1980-1999")

#save plot
plt.savefig("salaries.png")

# Show plot
plt.show()

```

```python
fig, ax = plt.subplots()

bar = ax.bar(df_bar['title'],df_bar['salary'])

def gradientbars(bars):
    grad = np.atleast_2d(np.linspace(0,1,256)).T
    ax = bars[0].axes
    lim = ax.get_xlim()+ax.get_ylim()
    for bar in bars:
        bar.set_zorder(1)
        bar.set_facecolor("none")
        x,y = bar.get_xy()
        w, h = bar.get_width(), bar.get_height()
        ax.imshow(grad, extent=[x,x+w,y,y+h], aspect="auto", zorder=0)
    ax.axis(lim)

gradientbars(bar)
# Create Ticks for Bar Chart's x_axis
plt.xticks(ticks, x_axis, rotation="50")

plt.ylabel("Salary in Dollars")
# plt.xlabel("Title")
plt.title("Salary/Title at Pewlett Hackard 1980-1999")
plt.annotate('Hmmmmm....', xy=(5,.5),  xycoords='data',
            xytext=(1.4, 0.2), textcoords='axes fraction',
            horizontalalignment='right', verticalalignment='bottom',
            )


#save plot
plt.savefig("salary_clarification.png")

# Show plot
plt.show()


```
